```python
## Estimation Statistics - 
# It is observed numerical value use to estimate an unknown population parameter .
# Types - Point estimation & Interval Estimate 

# Importance -  
#Estimation statistics involves using sample data to estimate population parameters. Point estimates provide a single value estimate, while interval estimates provide a range of values within which the true parameter is likely to lie, along with a certain level of confidence.
```

```python
# Write a Python function to estimate the population mean using a sample mean and standard deviation.=
import scipy.stats as stats
import math

def estimate_population_mean(sample_mean, sample_std, sample_size, confidence_level):

    margin_of_error = stats.t.ppf((confidence_level + 1) / 2, df=sample_size - 1) * (sample_std / math.sqrt(sample_size))
    
    
    lower_bound = sample_mean - margin_of_error
    upper_bound = sample_mean + margin_of_error
    
    return lower_bound, upper_bound

sample_mean = 78.5
sample_std = 4.2
sample_size = 30
confidence_level = 0.95

confidence_interval = estimate_population_mean(sample_mean, sample_std, sample_size, confidence_level)
print(f"Confidence Interval: {confidence_interval}")
```

```python
## Hypothesis testing - 
# Hypothesis testing is a statistical method used to make inferences(conclusion) about population parameters based on a sample of data. 
## Importance - 
# Hypothesis testing is a crucial tool in statistics that enables researchers and decision-makers to draw conclusions about populations based on sample data, fostering informed decision-making and advancing scientific understanding.
```

```python
### Create a hypothesis that states whether the average weight of male college students is greater than the average weight of female college students.
# Asum = WMa = Average Weight of male student
# Asum = WFa =  Average Weight of male student

#1)Null Hypothesis - Their is no any differance between WMa & WFa

#2)Alternate Hypothesis - Asum WMa is Greater than the WMa is Greater than WFa.

#3)Why Asum WMa is Greater than the WMa is Greater than WFa- As per data after using Statistical methods and Anylysis Proved the WMa is Greater than WFa.

#4)Hence We reject the null anylysis & ait is alternative anylysis.
```

```python
###Write a Python script to conduct a hypothesis test on the difference between two population means,given a sample from each population.
import numpy as np
from scipy.stats import ttest_ind

def conduct_t_test(sample1, sample2, alpha=0.05):
    t_statistic, p_value = ttest_ind(sample1, sample2, equal_var=False)

    if p_value < alpha:
        print("There is enough evidence to suggest a significant difference.")
    else:
        print("There is not enough evidence to suggest a significant difference.")

    return t_statistic, p_value

sample_male = np.array([76,59, 85, 78, 72])
sample_female = np.array([75, 60, 59, 68, 62])

t_stat, p_val = conduct_t_test(sample_male, sample_female)
print(f"T-statistic: {t_stat}")
print(f"P-value: {p_val}")
```

```python
##Null Hypothesis-
#The null hypothesis is a statement that there is no significant difference or effect. It often represents the status or a default assumption.
#Example - The average height of male and female students in a school is the same

##Alternative Hypothesis-
#The alternative hypothesis is a statement that contradicts the null hypothesis. It represents what the researcher is trying to find evidence .
#Example - The average height of male students is greater than the average height of female
```

```python
import numpy as np
from scipy.stats import ttest_ind

def two_sample_t_test(sample1, sample2, alpha=0.05):
   
    t_statistic, p_value = ttest_ind(sample1, sample2)

    if p_value < alpha:
        print("Reject the null hypothesis. There is enough evidence to suggest a significant difference.")
    else:
        print("Fail to reject the null hypothesis. There is not enough evidence to suggest a significant difference.")

    return t_statistic, p_value


sample1 = np.array([23, 25, 28, 32, 30])
sample2 = np.array([18, 20, 22, 26, 24])

t_stat, p_val = two_sample_t_test(sample1, sample2)
print(f"T-statistic: {t_stat}")
print(f"P-value: {p_val}")
```

```python
import scipy.stats as stats
import math

# Given values
sample_mean = 500
standard_deviation = 50
sample_size = 50
confidence_level = 0.95

# Calculate the Z-score for the 95% confidence level
z_score = stats.norm.ppf((1 + confidence_level) / 2)

# Calculate the margin of error
margin_of_error = z_score * (standard_deviation / math.sqrt(sample_size))

# Calculate the confidence interval
lower_bound = sample_mean - margin_of_error
upper_bound = sample_mean + margin_of_error

print(f"95% Confidence Interval for Population Mean: (${lower_bound:.2f}, ${upper_bound:.2f})")
```
