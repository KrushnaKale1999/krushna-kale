```python
## Calculate the 95% confidence interval for a sample of data with a mean of 50 and a standard deviation of 5 using Python. Interpret the results.
import scipy.stats as stats
import math

def cal_confindence_interval(sample_mean , sample_std_dev ,sample_size,confidence_level=0.95):
    margin_error = stats.norm.ppf((1 + confidence_level) / 2)*(sample_std_dev/math.sqrt(sample_size))
    confidence_interval= (sample_mean-margin_error , sample_mean+margin_error)
    return confidence_interval
                                                               
sample_mean =50
sample_std_dev = 5
sample_size=25 #Asum
                                                               
confidence_interval= cal_confindence_interval( sample_mean,sample_size ,sample_std_dev ) 
print(f'95% confidence Interval Is: {confidence_interval}')                                                             
```

```python
###Conduct a chi-square goodness of fit test to determine if the distribution of colors of M&Ms in a bag 
   ##matches the expected distribution of 20% blue, 20% orange, 20% green, 10% yellow, 10% red, and 20 brown. Use Python to perform the test with a significance level of 0.05.

from scipy.stats import chisquare
import numpy as np 
import math

observed_counts = np.array([0.2, 0.2, 0.2, 0.1, 0.1, 0.2])

expected_counts = np.array([0.1, 0.2, 0.2, 0.2, 0.1, 0.2]) * sum (observed_counts)


chi2_stat, p_value = chisquare(f_obs= observed_counts, f_exp=expected_counts)

print(f"Chi-square statistic: {chi2_stat}")
print(f"P-value: {p_value}")

# Compare p-value to the significance level
alpha = 0.05
if p_value < alpha:
    print("Reject the null hypothesis. The distribution of colors is significantly different.")
else:
    print("Fail to reject the null hypothesis. The distribution of colors is not significantly different.")
```

```python
## A study of the prevalence of smoking in a population of 500 individuals found that 60 individuals smoked. Use Python to calculate the 95% confidence interval for the true proportion of individuals in thepopulation who smoke.

import statsmodels.api as sm

total_population = 500
smokers = 60

smoking_proportion = smokers / total_population

conf_interval = sm.stats.proportion_confint(smokers, total_population, alpha=0.05)
print(f"Sample proportion: {smoking_proportion}")
print(f"95% Confidence Interval: {conf_interval}")
```

```python
## Calculate the 90% confidence interval for a sample of data with a mean of 75 and a standard deviation of 12 using Python. Interpret the results.
import scipy.stats as stats
import math

def cal_confindence_interval(sample_mean , sample_std_dev ,sample_size,confidence_level=0.90):
    margin_error = stats.norm.ppf((1 + confidence_level) / 2)*(sample_std_dev/math.sqrt(sample_size))
    confidence_interval= (sample_mean-margin_error , sample_mean+margin_error)
    return confidence_interval
                                                               
sample_mean =75
sample_std_dev = 12
sample_size=60 #Asum
                                                               
confidence_interval= cal_confindence_interval( sample_mean,sample_size ,sample_std_dev ) 
print(f'90% confidence Interval Is: {confidence_interval}')                                                             
```

```python
## Use Python to plot the chi-square distribution with 10 degrees of freedom. Label the axes and shade thearea corresponding to a chi-square statistic of 15.
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import chi2

df = 10
x = np.linspace(0, 30, 1000)
y = chi2.pdf(x, df)

plt.plot(x, y, label=f'Chi-square Distribution (df={df})')
x_shaded = np.linspace(0, 15, 100)
y_shaded = chi2.pdf(x_shaded, df)
plt.fill_between(x_shaded, y_shaded, color='blue', alpha=0.3, label='Chi-square statistic of 15')
plt.xlabel('Chi-square Statistic')
plt.ylabel('Probability Density Function (PDF)')
plt.legend()
plt.title('Chi-square Distribution with Shaded Area')
plt.show()
```

```python
## A random sample of 1000 people was asked if they preferred Coke or Pepsi. Of the sample, 520preferred Coke. Calculate a 99% confidence interval for the true proportion of people in the population whoprefer Coke.
import statsmodels.api as sm

sample_size = 1000
coke_preference = 520

coke_proportion = coke_preference / sample_size

conf_interval = sm.stats.proportion_confint(coke_preference, sample_size, alpha=0.01)
print(f"Sample proportion: {coke_proportion}")
print(f"99% Confidence Interval: {conf_interval}")
```
