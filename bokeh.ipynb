```python
from bokeh.plotting import figure , output_file , show
output_file= ('test.html')
```

```python
x =[2,3,4,5,6]
y =[6,7,8,9,5]
```

```python
p = figure(title='line_plot')
p.circle(x,y)
show(p)
```

```python
#Glyphs can be points, lines, circles, rectangles, and other geometric shapes that visually convey information about the underlying data. Bokeh provides a variety of glyphs that you can use to customize and enhance your visualizations.
```

```python
p = figure(title='line_plot')
p.line(x,y)
show(p)
```

```python
p = figure(title='line_plot')
p.scatter(x,y ,fill_color='red',size=20,legend_label='red points')
show(p)
```
